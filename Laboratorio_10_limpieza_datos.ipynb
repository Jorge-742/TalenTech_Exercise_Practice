## Part 1.
Ejercicio de práctica 1.

Salud Digital y Análisis de Datos de Pacientes
Se tiene un conjunto de datos sobre pacientes que incluye nombre, edad, fecha de ingreso y resultados de pruebas. Algunos registros están incompletos o contienen errores.


```python
# Datos
data = {'Nombre': ['Carlos', 'Ana', 'Luis', 'Pedro', 'María', None],
        'Edad': [34, 29, None, 45, 25, 33],
        'Fecha_ingreso': ['2021-01-15', '2020/03/12', '2022-07-01', '2021/12/01', '2021-05-20', '2020-03-12'],
        'Resultado_prueba': ['Positivo','Negativo', 'Positivo', 'Positivo', 'Negativo','Negativo']}
```

```python
# Creación del dataframe
df = pd.DataFrame(data)
df.head(10)
```

```python
# Limpieza de datos

#Dataframe original
print(df)

#Eliminación de datos faltantes en la columna nombre
print('\n Eliminicación de datos faltantes en la columna "Nombre"')
df = df.dropna(subset=['Nombre'])
print(df)

# Imputación de valores faltantes en la columna edad con el promedio de edad
print('\n Imputación de valores faltanes en la columna "Edad"')
df['Edad'] = df['Edad'].fillna(df['Edad'].mean())
print(df)

# Formateo de fecha
print('\n Formateo de fecha')
df['Fecha_ingreso'] = pd.to_datetime(df['Fecha_ingreso'].str.replace('/', "-"))
print(df)

```

**Ejercicio 2.**

Comercio Electrónico y Recomendaciones Personalizadas
Se tiene un conjunto de datos sobre transacciones de clientes que incluye el ID del cliente, el producto comprado, la cantidad, el precio y la fecha de la transacción. Algunos registros tienen errores en los precios y fechas.




```python
# Datos

datos = {'Cliente ID': [1, 2, 3, 4, 5],
         'Producto' : ['A', 'B', 'C', 'D', 'E'],
         'Cantidad' : [10, 20, None, 5, 15],
         'Precio' : ['100', '200$','300', 'cuatrocientos','500'],
         'Fecha Transacción' : ['2023-01-15', '2023/02/20','2023-03-25','2023/04/30', '2023-05-10']}

df2 = pd.DataFrame(datos)
df2.head(10)
```

```python
# Limpieza de datos

# 1. dataframe original
print('\n DataFrame original')
print(df2)

# 2. Imputación de valores faltantes
print('\n Imputación de valores faltantes')
df2['Cantidad'] = df2['Cantidad'].fillna(df2['Cantidad'].median())
print(df2)

# 3. correción de formato, remplazo de texto no numérico
# Conversión de tipo de dato, eliminación de caracteres no númericos.
print('\n correción de formato de Precio')
df2['Precio'] = df2['Precio'].str.replace('cuatrocientos', '400').str.replace(r'[^0-9]', '', regex=True).astype(float)
print(df2)

# 4. correción formato fecha.
print('\n Formateo de fecha')
df2['Fecha Transacción'] = pd.to_datetime(df2['Fecha Transacción'].str.replace('/', "-"))
print(df2)
```

**Ejercicio 3**

**Escenario:** eres un analista de datos en una pequeña empresa que está recopilando información sobre sus clientes.
El equipo de ventas te entrega un archivo con los siguientes datos:
- Nombre de los clientes
- Edad
- Correo electrónico
- País
- Género
- Número de visitas a la tienda

Sin embargo, cuando abres el archivo, te das cuenta de que hay problemas en los datos:
1. Valores faltantes: Algunos clientes no tienen información sobre su edad o correo electrónico.
2. Duplicados: Algunos clientes han sido registrados más de una vez.
3. Inconsistencias: Hay diferentes formas de escribir el género ("M", "Masculino", "F", "Femenino").
4. Errores tipográficos: El nombre de algunos países está en minúsculas o mal escrito.
5. Valores inválidos: Hay una edad registrada que parece imposible (200 años).

```python
data = """Nombre,Edad,Email,Pais,Genero,Visitas
Juan,25,juan@mail.com,colombia,M,5
Maria,30,maria@mail.com,Colombia,F,7
Pedro,,pedro@mail.com,Mexico,H,6
Ana,29,ana@mail.com,Argentina,F,4
Luisa,28,luisa@mail.com,Colombia,Femenino,8
Carlos,200,carlos@mail.com,Brazil,M,2
Juan,25,juan@mail.com,colombia,M,5"""

#Creacción de diccionario
lines = data.strip().split('\n')
header = lines[0].split(',')
rows = [line.split(',') for line in lines[1:]]
data_dict = {header[i]: [row[i] for row in rows] for i in range(len(header))}

# Creacción de dataframe
df = pd.DataFrame(data_dict)
df.head(10)
```

```python
# Limpieza de datos

# 0 conversion a tipos de datos
df['Edad'] = pd.to_numeric(df['Edad'], errors='coerce')
df['Visitas'] = pd.to_numeric(df['Visitas'], errors='coerce')
print('\n Conversion de tipos de datos')
print(df)
```

```python
#1. Eliminación de duplicados
print('\n Eliminación de duplicados')
df = df.drop_duplicates()
print(df)
```

```python
# 2 correción de valores
print('\n Correción de valores')
df.loc[df['Edad'] > 100, 'Edad'] = df['Edad'].mean()
print(df)
```

```python
# 3. relleno de valores faltantes
df['Edad'] = df['Edad'].fillna(df['Edad'].mean())
print('\n Relleno de valores faltantes')
print(df)
```

```python

# 4. Estandarización de Columnas

# 1. estandarización de la columna Email
print('\n Estandarización de columna Email')
df['Email'] = df['Email'].str.lower()
print(df)

# 2. Estandarización columna país
print('\n Estandarización de columna País')
df['Pais'] = df['Pais'].str.capitalize().str.replace('Brazil', 'Brasil')
print(df)

# 3. Estandarización columna género
print('\n Estandarización de columna Género')
df['Genero'] = df['Genero'].str.replace('M', 'Masculino').str.replace('F', 'Femenino').str.replace('H', 'Masculino').str.replace('Femeninoemenino', 'Femenino')
print(df)



```

```python
#almanecamiento de la limpieza de datos en una nueva variable.
df_final = df

print('\n DataFrame final')
print(df_final)
df_final.to_csv('df_final.csv', index=False)
df_final.dtypes
```
